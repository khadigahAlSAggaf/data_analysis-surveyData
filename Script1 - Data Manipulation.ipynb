```python
id_vars=list(dataset_modified.columns)[:8] # from column 0 to 7
value_vars=list(dataset_modified.columns)[8:] # from column 8 to end
```

```python
dataset_melted=dataset_modified.melt(id_vars=id_vars,value_vars=value_vars, var_name="Quesion + Subquestion", value_name="Answer")
#dataset_melted
dataset_melted
```

```python
dataset_merged = pd.merge(left=dataset_melted, right=questions, how="left",
                          left_on="Quesion + Subquestion", right_on="Question + Subquestion")
```

```python
print("Origenal data len=", len(dataset_melted))
print("Merged data len=", len(dataset_merged))

#dataset_merged
```

```python
respondes=dataset_merged[dataset_merged["Answer"].notna()]
```

```python
respondes
```

```python
respondes = respondes.groupby(
    "Question")["Respondent ID"].nunique().reset_index()
#respondes.rename(columns={"Respondent ID": "Respondent"}, inplace=True)
respondes
```

```python
dataset_merged_two = pd.merge(left=dataset_merged, right=respondes, how="left",
                         left_on="Question", right_on="Question")
print("Merged data len=", len(dataset_merged))
print("Merged_two data len=", len(dataset_merged_two))
#dataset_merged_two
```

```python
same_answer=dataset_merged#[dataset_merged["Answer"].notna()]
same_answer=same_answer.groupby(["Quesion + Subquestion","Answer"])["Respondent ID"].nunique().reset_index()
same_answer.rename(columns={"Respondent ID":"Same Answer"}, inplace=True)
#same_answer
```

```python
dataset_merged_three = pd.merge(left=dataset_merged_two, right=same_answer, how="left",
                                left_on=["Quesion + Subquestion", "Answer"], right_on=["Quesion + Subquestion", "Answer"])
dataset_merged_three["Same Answer"].fillna(0, inplace=True)
print("Merged data len=", len(dataset_merged_two))
print("Merged_two data len=", len(dataset_merged_three))
#dataset_merged_three
```

```python
output=dataset_merged_three.copy()
output.rename(columns={"Identify which division you work in.-Response":"Division Primary",	"Identify which division you work in.-Other (please specify)":"Division Secondary",	"Which of the following best describes your position level?-Response":"Position",	"Which generation are you apart of?-Response":"Generation","Please select the gender in which you identify.-Response":"Gender","Which duration range best aligns with your tenure at your company?-Response":"Tenuer",	"Which of the following best describes your employment type?-Response":"Emplomynt type"}, inplace=True)
#output
```

```python
output.to_csv('final_output.csv', index=False)
```
